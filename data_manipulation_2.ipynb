```python
import pandas as pd

df= pd.read_csv('Pokemon.csv')
#for .txt file p.read_csv('pokemon.txt', delimeter ='\t')

df.head(5)
```

```python
#basic operations
df.columns       # all columns
df.columns[3]    # the one at 3rd index so 4th feature

df['Name'][0:5]  #until 5th rows at this column

df[['Name','Type 1', 'Type 2']] # to view selected columns 

df.iloc[1]       #to see specific index row 

df.iloc[0:3]    #also same with to df[0:3]

df.iloc[2,1]    #to get specific value(index)
```

```python
#it is based on tex or numerical information more then index
# it is like where condition in SQL
df.loc[df['Type 1'] =='Fire'][0:10]
#if i add [0:3] at the end, it will give first 3 of them
```

```python
#describe gives all all mean, count,std,max... about each column
df.describe()
```

```python
#this sorts it like order by 
df.sort_values(['Name','HP'], ascending = False)
#if i want 1st one asc 2nd one des then 'ascending = [1,0]'
```

```python
#adding a column by using other columns
df['Total'] = df['HP'] + df['Attack']+ df['Defense']
df.head(5)
# drop acolumns, just make sure df = df.drop...
df = df.drop(columns = ['Total'])
df.head(5)
```

```python
#adding column again by using indexes

df['Total'] = df.iloc[:,4:10].sum(axis=1) # axis =1 means sum the rows, 0 means columns
df.head(5)
```

```python
# to reorder columns in df
col = list(df.columns)
df = df [col [0:4]+ [col[-1]] + col[4:12]]

df.head(5)
```

```python
#saving that file to same directory
df.to_csv('poke_ordered.csv',index = False)
#for .txt just add  sep = '\t'
```

```python
#filtering
filtered_df =df.loc[(df['Type 1']  == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] >60)] # or is like |
```

```python
#old index are still there
#add inplace to not create again a df. just alter it
filtered_df.reset_index(drop = True,inplace = True)
filtered_df
```

```python
#I check the ones have 'Mega' in it

df.loc[df['Name'].str.contains('Mega')]
#the one not have 'mega'  add ~ like below
#df.loc[~df['Name'].str.contains('Mega')]
```

```python
#Here I checked the ones who has fire or Grass in Type 1 with regex
import re

df.loc[df['Type 1'].str.contains('Fire|Grass' , regex =True)]
#to ignore capital letters just add flags=re.I in parantesis
# same with    df.loc[(df['Type 1']  == 'Grass') | (df['Type 1'] == 'Fire')]
```

```python
#to get the ones which starts with Cha

df.loc[df['Name'].str.contains('^Cha[a-z]*', flags=re.I, regex =True)]
# * mean 0 or more letter , ^ means starts with
```

```python
#This can be used for flagging. 
#I can check the criteria on Name and add column  with 1 hust changing 'Type 1'
df.loc[df['Type 1'] == 'Fire', 'Type 1']= 'Flamer'
df
```

```python
df.loc[df['Type 1'] == 'Flamer', 'Type 1']= 'Fire'
df
# This is the mix of regex and flag
# df.loc[df['Name'].str.contains('^T[a-z]*', flags=re.I, regex =True), 'Type 1']= 'Efsaneess'
```

```python
# Conditional Changes
# I created 2 columns after checking total>500.
# flag method 

df.loc [ df['Total'] > 500, ['Test1', 'Test2']] = ['Test 1', 'Test 2']
df
```

```python
df= df.drop(columns= ['Test1', 'Test2'])
```

```python
# Group by for aggregate
df.groupby(['Type 1']).mean().sort_values('HP', ascending=False)
#df.groupby(['Type 1']).count ()
#df.groupby(['Type 1']).sum ()
df['Count'] = 1
df.groupby(['Type 1']).count()['Count']  # to see total number of each group in type 1 
# I gave flag on count column
# this is 2 group by 
df.groupby(['Type 1', 'Type 2']).count()['Count'] 
```

```python
df = df.drop(columns = ['Count'])
```

```python
#this is for big datasets to get just 100 row and process each of them by their own and 
#bring them together in a df
new_df = pd.DataFrame (columns = df.columns)
for df in pd.read_csv('Pokemon.csv', chunksize=100):
    results = df.groupby (['Type 1']).count()
    new_df = pd.concat([new_df, results])
new_df.head(20)
    
    
```
