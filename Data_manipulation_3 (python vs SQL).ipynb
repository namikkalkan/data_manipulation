```python
#question 2 
#get the average number of HP and Attack for each Type1 and Type2

# Sql 
# select city, property, avg(bedrooms), avg(bathrooms) from property_datasets
# group by city, property

#Python
import pandas as pd
df= pd.read_csv('Pokemon.csv')
df.head(5)

#python code is like
df.groupby(['Type 1','Type 2'])[['HP','Attack']].mean().reset_index()
df.groupby(['Type 1','Type 2']).mean()[['HP','Attack']].reset_index()
```
